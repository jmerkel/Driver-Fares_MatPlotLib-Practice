```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
battleDF = pd.read_csv("Resources/got.csv")
battleDF.head()
```

```python
#Clear Non-King Battles

# Get attacker and defender data
attackerKing = battleDF.groupby(["attacker_king"]).count()["name"]
attackerKing
```

```python
defenderKing = battleDF["defender_king"].value_counts() 
# Works on a single column
defenderKing
```

```python
# Get total battle data
totalBattleDataDF = attackerKing.add(defenderKing, fill_value=0)
totalBattleDataDF
```

```python
# Configure plot and ticks
totalBattleDataDF.plot(kind="bar", color = "red")
```
